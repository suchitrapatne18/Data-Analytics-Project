## **Phase 1 — Handle Missing Values**

```python
# See which column has nulls
df['CouponCode'].isnull().sum()  # You'll see 309
```

```python
# Fill nulls with 'NO_COUPON' (means customer didn't use any coupon)
df['CouponCode'] = df['CouponCode'].fillna('NO_COUPON')

# Verify it worked
df['CouponCode'].isnull().sum()  # Should now be 0
```

## **Phase 2 — Remove Duplicates**

```python
# Check for fully duplicate rows
print("Duplicate rows:", df.duplicated().sum())

# Check for duplicate OrderIDs
print("Duplicate OrderIDs:", df['OrderID'].duplicated().sum())
```

```python
# If any duplicates exist, remove them like this:
df = df.drop_duplicates()

# For duplicate OrderIDs specifically:
df = df.drop_duplicates(subset='OrderID')
```

## **Phase 3 — Fix Formatting**

### **3a — Standardise Dates to YYYY-MM-DD:**

```python
# Convert to proper datetime first, then format as string
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%m-%d')

# Check it worked
df['Date'].head()
```

### **3b — Round prices to 2 decimal places:**

```python
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)
```

### **3c — Strip whitespace from all text columns:**

```python
# Loop through every text column and strip spaces
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
```

## **Step 4 — Export the Cleaned File**

```python
df.to_excel('Cleaned_Dataset.xlsx', index=False)
print("Done! File saved.")
```
